```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt(r"C:\Users\Gargi\Downloads\Ml\data.txt", delimiter=',')
```

```python
x_train=data[:,0]
y_train=data[:,1]
```

```python
x_train=(x_train-np.mean(x_train))/np.std(x_train)
y_train=(y_train-np.mean(y_train))/np.std(y_train)
```

```python
plt.scatter(x_train,y_train)
plt.xlabel("Area in Square feet")
plt.ylabel("Price in USD")
plt.title("Housing Prices vs Area")
plt.show()
```

```python
def compute_cost(x, y, w, b):
    m=x.shape[0] 
    cost=0
    for i in range(m):
        f_wb=w*x[i]+b
        cost=cost+(f_wb-y[i])**2
    total_cost=1/(2*m)*cost
    return total_cost
```

```python
def compute_gradient(x,y,w,b):
    m=x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb=w*x[i]+b 
        dj_dw_i=(f_wb-y[i])*x[i] 
        dj_db_i=f_wb-y[i] 
        dj_db+=dj_db_i
        dj_dw+=dj_dw_i 
    dj_dw=dj_dw/m 
    dj_db=dj_db/m 
        
    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,num_iter,cost_function,gradient_function):
    w=copy.deepcopy(w_in)
    j_history=[]
    p_history=[]
    w=w_in
    b=b_in
    for i in range(num_iter):
        dj_dw,dj_db=gradient_function(x,y,w,b)
        b=b-alpha*dj_db
        w=w-alpha*dj_dw
        if i<100000:
            j_history.append(cost_function(x,y,w,b))
            p_history.append([w,b])
    return w,b,j_history,p_history 
```

```python
w_init=0
b_init=0
iterations=1000
temporary_alpha=1.0e-2
w_final,b_final,J_hist,p_hist=gradient_descent(x_train,y_train,w_init,b_init,temporary_alpha,iterations,compute_cost,compute_gradient)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,constrained_layout=True,figsize=(12,4))
ax1.plot(J_hist[:100])
ax2.plot(1000+np.arange(len(J_hist[1000:])),J_hist[1000:])
ax1.set_title("Cost vs. iteration(start)");ax2.set_title("Cost vs. iteration (end)")
ax1.set_ylabel('Cost');
ax2.set_ylabel('Cost') 
ax1.set_xlabel('iteration step');
ax2.set_xlabel('iteration step')
plt.show()
```

```python
print(f"1000 sqft house prediction {w_final*1.0+b_final:0.1f} Thousand dollars")
print(f"1200 sqft house prediction {w_final*1.2+b_final:0.1f} Thousand dollars")
print(f"2000 sqft house prediction {w_final*2.0+b_final:0.1f} Thousand dollars")
```
